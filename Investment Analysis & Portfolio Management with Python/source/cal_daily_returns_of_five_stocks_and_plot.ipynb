```python
import pandas as pd
import requests
import numpy as np
from pandas_datareader import data
```

```python
def cal_return_plot(stock):
    # Only get the adjusted close.
    df = data.DataReader(stock, 
                           start='2017-1-1', 
                           end='2021-12-31', 
                           data_source='yahoo')
    print(df.head())
    
    df.rename(columns={'Date' : 'date', 'Adj Close': 'price_t' }, inplace=True)
    
    df['return'] = (df['price_t']/df['price_t'].shift(1)) - 1
    
    df['price_t'].plot(title=stock+' Adj. Closing Price',figsize=(12,8))
```

```python
cal_return_plot('MSFT')
```

```python
cal_return_plot('BBC')
```

```python
cal_return_plot('SKY')
```
