```python
import pandas as pd
import datetime 
```

```python
USD_data = pd.read_csv('./USD.csv')
binance_data = pd.read_csv('./binance.csv')
bitcoin_data = pd.read_csv('./bitcoin.csv')
ethereum_data = pd.read_csv('./ethereum.csv')
tether_data = pd.read_csv('./tether.csv')
tether_data.head()
```

```python
columns = ['Open', 'High','Low','Close','Volume' ]
USD_data.drop(columns, axis=1, inplace=True)
binance_data.drop(columns, axis=1, inplace=True)
bitcoin_data.drop(columns, axis=1, inplace=True)
ethereum_data.drop(columns, axis=1, inplace=True)
tether_data.drop(columns, axis=1, inplace=True)
USD_data
```

```python
crypto_predicated_data_1 = pd.merge(USD_data,ethereum_data, how='inner',on ='Date')
crypto_predicated_data_2 = pd.merge(crypto_predicated_data_1,binance_data, how='inner',on ='Date')
crypto_predicated_data_3 = pd.merge(crypto_predicated_data_2, bitcoin_data, how='inner',on ='Date')
crypto_predicated_data_4 = pd.merge(crypto_predicated_data_3,tether_data, how='inner',on ='Date')
```

```python
crypto_predicated_data_4.columns = ['Date','USD','ethereum','binance','bitcoin','tether']
```

```python
crypto_predicated_data_4.set_index('Date',inplace=True)
```

```python
crypto_predicated_data_4.to_csv('crypto_historical_close_price.csv')
```
